Basic Iterative Functions (lin, quad, log, sig)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#straight line [y = mx + b]

ax = plt.gca()
b = 30
m = 0.5
x = np.array(ax.get_xlim())
y = b + m * x
plt.plot(x, y, '--')
plt.show()
```

```python
#quadratic function [y = (x - m)^2 + b]

ax_quad = plt.gca()
b_quad = 20
m_quad = 25
x_quad = np.linspace(-1000, 1000, 100000)
y_quad = b + (x_quad - m_quad)**2
plt.plot(x_quad, y_quad)
plt.show()
```

```python
#log function [y = -log(x) and y = -log(1 - x)]

#ax_log = plt.gca()
x_log = np.arange(-50.5, 50.5, .25)

y1 = np.log10(abs(x_log)) * (-1)
y2 = np.log10(abs(1 - x_log)) * (-1)

plt.plot(x_log,y1)
plt.plot(x_log,y2)

plt.show()
```

```python
#sigmoid function [y = 1/(1 + e^-x)]

sigmoid = lambda x_sig: 1/(1 + np.exp(-x_sig))
x_sig = np.arange(-50.5, 50.5, 0.25)
#y_sig = np.linspace(0, len(x_sig), len(x_sig))
plt.plot(x_sig, sigmoid(x_sig))
plt.show()
```
